# Listado de silencios (principio y fin)

```python
from pydub import AudioSegment,silence
import pathlib
for file in pathlib.Path('./').glob('*.wav'):
	myaudio = intro = AudioSegment.from_wav(file)
	myaudio = myaudio[:20000]
	dBFS=myaudio.dBFS
	silences = silence.detect_silence(myaudio, min_silence_len=1000, silence_thresh=dBFS-16)
	silences = [((start/1000),(stop/1000)) for start,stop in silences]
	print(str(file) + " " + str(silences))
```

# Listado de silencios (segundos de los silencios)

```python
from pydub import AudioSegment,silence
import pathlib
for file in pathlib.Path('./').glob('*.wav'):
	myaudio = intro = AudioSegment.from_wav(file)
	myaudio = myaudio[:20000]
	dBFS=myaudio.dBFS
	silences = silence.detect_silence(myaudio, min_silence_len=1000, silence_thresh=dBFS-16)
	silences = [round((stop/1000)-(start/1000),2) for start,stop in silences]
	print(str(file) + " " + str(silences))
```

# Listado de silencios (media de llamada)

```python
from pydub import AudioSegment,silence
import statistics
import pathlib
for file in pathlib.Path('./').glob('*.wav'):
	myaudio = intro = AudioSegment.from_wav(file)
	#myaudio = myaudio[:20000]
	dBFS=myaudio.dBFS
	silences = silence.detect_silence(myaudio, min_silence_len=1000, silence_thresh=dBFS-16)
	silences = [round((stop/1000)-(start/1000),2) for start,stop in silences]
	print(str(file) + " " + str(statistics.mean(silences)))
```

# Gr√°ficas de audio

```python
import wave
import numpy as np
import matplotlib.pyplot as plt



for file in pathlib.Path('./').glob('*.wav'):
    signal_wave = wave.open(file.name, 'r')
    sample_rate = 10000000
    sig = np.frombuffer(signal_wave.readframes(sample_rate), dtype=np.int16)
    sig = sig[:]
    plt.figure(1)
    plot_a = plt.subplot(211)
    plot_a.plot(sig)
    plot_a.set_xlabel(file.name)

    plt.show()
```
